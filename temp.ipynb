```python
import numpy as np
from sklearn.linear_model import Lasso

# 生成示例數據
np.random.seed(0)
X = np.random.rand(100, 10)  # 100個樣本，10個特徵
y = X.dot(np.random.rand(10)) + np.random.randn(100)  # 目標變量

# 初始化Lasso模型並設置alpha
lasso = Lasso(alpha=0.1)

# 訓練模型
lasso.fit(X, y)
print(X.shape)
print(y.shape)
# 獲取係數和截距
coefficients = lasso.coef_
intercept = lasso.intercept_

# 輸出結果
print("係數:", coefficients)
print("截距:", intercept)

# 使用模型進行預測
predictions = lasso.predict(X)
print("預測值:", predictions)
```
